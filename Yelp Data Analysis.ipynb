CREATE TABLE Business (
    business_id NVARCHAR(50) PRIMARY KEY,
    name NVARCHAR(255),
    city NVARCHAR(100),
    state NVARCHAR(50),
    stars FLOAT,
    is_open INT,
);

CREATE TABLE Category (
    category_id INT IDENTITY(1,1) PRIMARY KEY,
    business_id NVARCHAR(50),
    category_name NVARCHAR(255),
    FOREIGN KEY (business_id) REFERENCES Business(business_id)
);

CREATE TABLE Review (
    review_id NVARCHAR(50) PRIMARY KEY,
    user_id NVARCHAR(50),
    business_id NVARCHAR(50),
    stars INT,
    date DATETIME,
    text NVARCHAR(MAX),
    FOREIGN KEY (user_id) REFERENCES Users(user_id),
    FOREIGN KEY (business_id) REFERENCES Business(business_id)
);

CREATE TABLE Tip (
    user_id NVARCHAR(50),
    business_id NVARCHAR(50),
    text NVARCHAR(MAX),
    date DATETIME,
    PRIMARY KEY (user_id, business_id),
    FOREIGN KEY (user_id) REFERENCES Users(user_id),
    FOREIGN KEY (business_id) REFERENCES Business(business_id)
);

CREATE TABLE Checkin (
    business_id NVARCHAR(50) PRIMARY KEY,
    date NVARCHAR(MAX),
    FOREIGN KEY (business_id) REFERENCES Business(business_id)
);

##Milestone3: Extract Business Insights
We are planning to open a new restaurant and used the Yelp dataset to analyze market opportunities.
Our goal is to determine where and what type of restaurant to open based on existing competition, customer preferences, and market saturation.

**Market Analysis**
Where is the competition? How saturated is the market? What are competitive ratings?

```python
# Q1: Which city has the most restaurants?
import pandas as pd
query = """
SELECT city, COUNT(*) AS restaurant_count
FROM Business
GROUP BY city
ORDER BY restaurant_count DESC;
"""

df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Reveals highly competitive markets.
```

```python
# Q2: Which restaurant categories are most common in each state?
query = """
SELECT state, category_name, category_count
FROM (
    SELECT
        b.state,
        c.category_name,
        COUNT(*) AS category_count,
        ROW_NUMBER() OVER (PARTITION BY b.state ORDER BY COUNT(*) DESC) AS rn
    FROM Business b
    JOIN Category c ON b.business_id = c.business_id
    GROUP BY b.state, c.category_name
) ranked
WHERE rn <= 2
ORDER BY state, category_count DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(30))

#Insight: Understand regional cuisine preferences and which markets are saturated with similar concepts.
```

```python
# Q3: Which cities have fewer restaurants but high average ratings?
query = """
SELECT state, AVG(stars) AS avg_rating
FROM Business
GROUP BY state
ORDER BY avg_rating DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Identifies states with consistently strong restaurant performance, suggesting regions with mature and quality-driven dining markets.
```

```python
# Q4: Top restaurant categories
query = """
SELECT category_name, COUNT(*) AS business_count
FROM Category
GROUP BY category_name
ORDER BY business_count DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Identifies the most common restaurant categories  helps understand dominant cuisines and market trends.
```

**Customer Insights**
What do customers value? What are common complaints? What drives success?

```python
# Q5: Which restaurant categories have the highest average ratings?
query = """
SELECT c.category_name, AVG(b.stars) AS avg_rating, COUNT(*) AS count
FROM Business b
JOIN Category c ON b.business_id = c.business_id
GROUP BY c.category_name
HAVING COUNT(*) > 10
ORDER BY avg_rating DESC;
"""

df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Shows which cuisines or restaurant types customers appreciate most, ideal for selecting a concept that aligns with customer preferences.
```

```python
# Q6: Which cities have the most different kinds of restaurants?
query = """
SELECT
    b.city,
    COUNT(DISTINCT c.category_name) AS category_variety
FROM Business b
JOIN Category c ON b.business_id = c.business_id
GROUP BY b.city
ORDER BY category_variety DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Show which cities have the most diverse restaurant options
```

```python
# Q7: Which categories in a specific state have few restaurants but high ratings?
query = """
SELECT
    c.category_name,
    COUNT(*) AS count,
    AVG(b.stars) AS avg_rating
FROM Business b
JOIN Category c
    ON b.business_id = c.business_id
WHERE b.state = 'LA'
GROUP BY c.category_name
HAVING COUNT(*) < 5
ORDER BY avg_rating DESC;
"""

df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Finding opportunites in a specific state
```

 **Opportunity Identification**

```python
# Q8: Which restaurant categories are popular but have declining ratings?
query = """
SELECT c.category_name, COUNT(*) AS count, AVG(b.stars) AS avg_rating
FROM Business b
JOIN Category c ON b.business_id = c.business_id
GROUP BY c.category_name
HAVING COUNT(*) > 20 AND AVG(b.stars) < 3.5
ORDER BY count DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Identifies crowded but underperforming segments where you can outperform competitors with better quality or experience.
```

```python
# Q9: Which restaurant categories have few competitors but high ratings?
query = """
SELECT c.category_name, COUNT(*) AS count, AVG(b.stars) AS avg_rating
FROM Business b
JOIN Category c ON b.business_id = c.business_id
GROUP BY c.category_name
HAVING COUNT(*) > 5
ORDER BY avg_rating DESC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Insight: Finds niche opportunities with low competition but high customer satisfaction.
```

```python
# Q10: Compare restaurant density vs. satisfaction
query = """
SELECT city, COUNT(*) AS restaurant_count, AVG(stars) AS avg_rating
FROM Business
GROUP BY city
HAVING COUNT(*) > 5
ORDER BY avg_rating DESC, restaurant_count ASC;
"""
df_result = pd.read_sql(query, engine)
display(df_result.head(10))

#Finds the high satisfaction, low competition place. Ideal for choosing new restaurant locations.
```
